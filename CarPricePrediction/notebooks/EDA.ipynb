```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"D:\OI\Task_3\notebooks\data\car data.csv")
```

```python
data.head()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set1_r')
```

```python
import warnings
warnings.filterwarnings("ignore")
years = data["Year"].unique()
print(f"Years :{sorted(years)}")
sns.pairplot(data = data)
```

```python
sns.scatterplot(data = data,x = 'Present_Price',y = 'Selling_Price',hue = 'Selling_type',palette = 'bwr')
```

```python
grp_year = pd.DataFrame(data.groupby(by=['Year'])['Car_Name'].count()).reset_index().rename(columns={'Car_Name': 'No of Cars Sold'})
plt.figure(figsize = (12,12))
sns.barplot(data = grp_year,x = 'Year',y = 'No of Cars Sold')
```

```python
numeric_data = data.select_dtypes(include='number')
sns.heatmap(data=numeric_data.corr(), annot=True, cmap='viridis')
plt.show()
```
